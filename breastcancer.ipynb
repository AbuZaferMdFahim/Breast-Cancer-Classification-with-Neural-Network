```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
# loading the dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()

# or
# breast_cancer_dataset = pd.read_csv('data.csv')
```

```python
# loading the data to dataframe
breast_cancer = pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
# adding the 'target' column to the data frame
breast_cancer['label'] = breast_cancer_dataset.target
```

Separating the features and target

```python
X = breast_cancer.drop(columns='label',axis=1)
Y = breast_cancer['label']
```

```python
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

Building the Nural Network

```python
# importing tensorflow and kerus
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras 
```

```python
# setting up the layers of Neural Network

# Increase model complexity
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(64, activation='relu'),  # Increase neurons
    keras.layers.Dropout(0.5),  # Add dropout for regularization
    keras.layers.Dense(32, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the nural network
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Training the neural nework
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=20)
```

Visualization accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training data', 'Validation data'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training data', 'Validation data'], loc='upper right')
plt.show()
```

Accuracy of model on test data

```python
# Evaluate the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print("Test Accuracy:", accuracy)
```

```python
Y_pred= model.predict(X_test_std)
```

Model predict gives the prediction of probablity of each class for that data point

```python
# converting the prediction probablity to class labels
Y_pred_labels = [np.argmax(i) for i in Y_pred ]
print(Y_pred_labels)
```
